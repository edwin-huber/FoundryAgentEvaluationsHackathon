# Session 6 Solution: Interpret Results and Apply Quality Gates

```python
from pathlib import Path
import json

summary_path = Path.cwd().parent / "artifacts" / "session6_sample_summary.json"
if not summary_path.exists():
    summary_path = Path.cwd() / "solutions" / "artifacts" / "session6_sample_summary.json"

summary = json.loads(summary_path.read_text(encoding="utf-8"))
summary
```

```python
thresholds = {"coherence": 0.85, "relevance": 0.85, "task_adherence": 0.85}
quality_gate = all(summary.get(metric, 0) >= target for metric, target in thresholds.items())
print("Quality gate pass:", quality_gate)
{"summary": summary, "thresholds": thresholds, "pass": quality_gate}
```
